    Selecione somente os imóveis classificados com tipo 'Apartamento'.

    Selecione os imóveis classificados com tipos 'Casa', 'Casa de Condomínio' e 'Casa de Vila'.

    Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites.

    Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00.


```python
import pandas as pd
```

```python
data = pd.read_csv('datas/residential_rent.csv', sep=';')
```

```python
data.head(10)
```

```python
apt_selection = data["Tipo"] == 'Apartamento'
apt_selection
```

```python
data_apt = data[apt_selection]
data_apt.head(10)
```

```python
data_apt.to_csv('datas/apt_rent.csv', sep=';', index=False)
```

```python
data_apt = pd.read_csv('datas/apt_rent.csv', sep=';')
data_apt.head(10)
```

```python
houses = [
    'Casa', 
    'Casa de Condomínio',
    'Casa de Vila'
]
houses_selection = data['Tipo'].isin(houses)
houses_selection.head(10)
```

```python
houses_data = data[houses_selection]
houses_data.head(10)
```

```python
houses_data.to_csv('datas/houses_csv', sep=';', index=False)
```

```python
houses_data = pd.read_csv('datas/houses_csv', sep=';')
```

```python
houses_data
```

```python
meters_selection = (data['Area'] >= 60) & (data['Area'] <= 180)
meters_selection
```

```python
meters_data = data[meters_selection]
meters_data.head(10)
```

```python
meters_data.to_csv('datas/meters_csv', sep=';', index=False)
meters_data = pd.read_csv('datas/meters_csv', sep=';')
```

```python
meters_data.head(10)
```

```python
query_selection = (data['Quartos'] >= 4) & (data['Valor'] <= 2000)
query_selection
```
