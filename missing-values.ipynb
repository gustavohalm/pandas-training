## Tratamento de dados faltantes

```python
import pandas as pd
```

```python
data = pd.read_csv('datas/residential_rent.csv', sep=';')
```

```python
data[data['Valor'].isnull()]
```

```python
data.dropna(subset = ['Valor'], inplace=True)
data.shape[0]
```

```python
selection_empty_apt = (data['Tipo'] == 'Apartamento') & (data['Condominio'].isnull())
empty_condo_apt = data[selection_empty_apt]
empty_condo_apt
```

```python
data.drop(empty_condo_apt.index)
```

```python
data = data[~selection_empty_apt]
data
```

```python
data.fillna(0, inplace=True) # OR {'Condominio': 0 , 'IPTU':0}
```

```python
data[data['IPTU'].isnull()]
```

```python
data.to_csv('datas/residential_rent.csv', sep=';', index=False)
```
